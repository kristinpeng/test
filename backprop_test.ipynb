```python
import numpy as np

# 1. Define activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    # x is already the output of the sigmoid function
    return x * (1 - x)

# 2. Define Dataset (XOR Gate)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0], [1], [1], [0]])

# Seed for reproducibility
np.random.seed(42)

# 3. Initialize Hyperparameters and Network Parameters
epochs = 10000
learning_rate = 0.1

input_size = 2
hidden_size = 3
output_size = 1

# Randomly initialize weights and biases
W1 = np.random.uniform(size=(input_size, hidden_size))
b1 = np.zeros((1, hidden_size))
W2 = np.random.uniform(size=(hidden_size, output_size))
b2 = np.zeros((1, output_size))

# 4. Training Loop
for epoch in range(epochs):
    # --- FORWARD PASS ---
    # Input to Hidden Layer
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)
    
    # Hidden Layer to Output Layer
    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2) # Final prediction
    
    # Calculate Mean Squared Error Loss
    loss = np.mean((y - A2) ** 2)

    # --- BACKPROPAGATION (BACKWARD PASS) ---
    # 1. Error and gradient at Output Layer
    error_output = A2 - y
    d_output = error_output * sigmoid_derivative(A2)
    
    # 2. Error and gradient at Hidden Layer
    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * sigmoid_derivative(A1)
    
    # 3. Calculate parameter gradients
    dW2 = np.dot(A1.T, d_output)
    db2 = np.sum(d_output, axis=0, keepdims=True)
    
    dW1 = np.dot(X.T, d_hidden)
    db1 = np.sum(d_hidden, axis=0, keepdims=True)
    
    # --- GRADIENT DESCENT (UPDATE WEIGHTS) ---
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    if epoch % 2000 == 0:
        print(f"Epoch {epoch:5d} | Loss: {loss:.4f}")

# Verify predictions after training
print("\nFinal Predictions:")
print(A2)
```

```python
def predict(new_input, W1, b1, W2, b2):
    """
    Runs a forward pass through the trained network to predict an outcome.
    """
    # Convert input list to a 2D numpy array row
    X_new = np.array([new_input])
    
    # Forward pass through hidden layer
    Z1 = np.dot(X_new, W1) + b1
    A1 = sigmoid(Z1)
    
    # Forward pass through output layer
    Z2 = np.dot(A1, W2) + b2
    raw_prediction = sigmoid(Z2)[0][0]
    
    # Convert probability to a hard 0 or 1 binary choice
    binary_prediction = 1 if raw_prediction > 0.5 else 0
    
    return binary_prediction, raw_prediction


# --- EXAMPLES OF USE ---

# Test a single custom input
test_data = [1, 0]
binary_res, raw_prob = predict(test_data, W1, b1, W2, b2)

print(f"Testing input: {test_data}")
print(f"Raw Probability: {raw_prob:.4f}")
print(f"Final Decision : {binary_res}")

# Interactive prompt loop
print("\n--- Try It Yourself ---")
try:
    in1 = int(input("Enter first binary digit (0 or 1): "))
    in2 = int(input("Enter second binary digit (0 or 1): "))
    
    final_choice, confidence = predict([in1, in2], W1, b1, W2, b2)
    print(f"The trained network says the XOR output is: {final_choice} (Confidence: {confidence:.2%})")
except ValueError:
    print("Please enter numbers only.")
```
